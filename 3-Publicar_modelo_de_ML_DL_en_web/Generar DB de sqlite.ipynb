#creando una base de datos SqlLite

```python
import sqlite3
import os

if os.path.exists('/content/drive/MyDrive/IA/PLN/DB_sentimientos_Esp.sqlite'):
    os.remove('/content/drive/MyDrive/IA/PLN/DB_sentimientos_Esp.sqlite')

conn = sqlite3.connect('/content/drive/MyDrive/IA/PLN/DB_sentimientos_Esp.sqlite')
c = conn.cursor()
c.execute('CREATE TABLE sentimientos_db (texto TEXT, sentimento INTEGER, fecha TEXT)')

example1 = 'que aburrido es estar en cuarentena…'
c.execute("INSERT INTO sentimientos_db (texto, sentimento, fecha) VALUES (?, ?, DATETIME('now'))", (example1, 0))

example2 = 'Estoy feliz de estar con mi familia'
c.execute("INSERT INTO sentimientos_db (texto, sentimento, fecha) VALUES (?, ?, DATETIME('now'))", (example2, 1))

conn.commit()
conn.close()
```

#consultando una base de datos sqlLite

```python
conn = sqlite3.connect('/content/drive/MyDrive/IA/PLN/DB_sentimientos_Esp.sqlite')
c = conn.cursor()

c.execute("SELECT * FROM sentimientos_db  WHERE fecha BETWEEN '2022-01-01 10:10:10' AND DATETIME('now')")
results = c.fetchall()
conn.close()

print(results)
```
