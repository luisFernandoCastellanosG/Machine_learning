#SVM - Bill Authentication

##P1. importar librerias

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from  sklearn.svm import SVC
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as pl
import seaborn as sb
#guardar las graficas directamente en el libro
%matplotlib inline
```

##P2. Cargar datos desde GITHUB

```
df = pd.read_csv('https://raw.githubusercontent.com/luisFernandoCastellanosG/Machine_learning/master/Databasets/bill_authentication.csv')
df.head(10)
```

##P3. Conociendo los datos

```
#nombres de columnas
df.columns
#mostrar estadisticas del dataset
df.describe()
#conocer valores nulos
df.isnull().sum()
```

```
corr_matrix= df.corr().round(2)
sb.heatmap(data=corr_matrix,annot=True)
```

```
sb.set(style="whitegrid", context="notebook")
plot_data= df.columns
sb.pairplot(df[plot_data],height=2)
pl.tight_layout()
pl.show()
```

##P4. separando los datos (X y Y)

```
X= df.drop('Class',axis=1)
Y= df['Class']
```

##P5. separando en TRAIN y TEST

```
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.3, random_state=42)
```

##P6. instanciar algoritmo y entrenar

---
varios tipos de kernel:
*   linear (2 grupos)
*   Sigmoidal (clasificaciones binarias)
*   polinómico (varios grupos)
*   gaussiano radial ( separación por grupo tipo k-means)






```
svm = SVC(kernel='linear')

svm.fit(X_train,Y_train)
```

##P7. realizar predicciones

```
Y_pred = svm.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print(' Accuracy bill authentication SMV : '+ str(accuracy*100))
```
